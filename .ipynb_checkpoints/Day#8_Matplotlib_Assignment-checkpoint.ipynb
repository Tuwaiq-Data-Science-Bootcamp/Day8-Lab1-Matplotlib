```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
plt.plot([1,2,0,4,5]);
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(np.arange(1,5), labels=['1','2','3', '4'], explode = [.1,.1,.1,.1]);
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x = np.random.randn(500)
y = np.random.randn(500)
plt.scatter(x,y,alpha = 0.4, s = 100);
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
x = np.random.randn(500);
y = np.random.randn(500);
plt.scatter(x,y,alpha = 0.4, s = 100);

plt.title('Years', fontsize = 16, loc='left', color='r');
plt.xlabel('Days');
plt.ylabel('Sales');
```

## Q6: Create subplots with three charts and figsize=20,10

## plt.figure(figsize=[20,10])

```python
plt.suptitle('Subplots',fontsize = 16)

#first
plt.subplot(2,2,1) 
plt.pie(np.arange(1,5),labels=['Jan','Feb','Mar', 'Apr']);
plt.title('Months')

#second
plt.subplot(2,2,2)
plt.plot([1,2,0,4,5])
plt.plot([10,9,8,6,3])
plt.legend(['2008','2009'], shadow = True, loc = 0)
plt.title('Years');

#third
plt.subplot(2,2,3)
x = np.arange(0, 100);
y = np.random.randint(5, size=100);
plt.plot(x,y);
plt.legend(['2008','2009'], shadow = True, loc = "upper left");
```
