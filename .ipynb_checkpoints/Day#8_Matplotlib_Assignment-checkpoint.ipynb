```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
plt.plot([1,2,3,4,5],[10,20,3,1,5]);
plt.plot(pd.DataFrame([5,2,5,9]));
```

## Q2: Create pie chart with equal explode values.

```python
x= [1,1,1,1,1];
plt.pie(x);
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(np.random.randn(20),np.random.randn(20),alpha = 0.4 , s=50);
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(np.random.randn(20),np.random.randn(20),alpha = 0.4 , s=50);
plt.title('MMM', fontsize = 16, loc='left', color='r');
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.plot([1,2,3,4,5],color ='blue');
plt.plot(pd.DataFrame([5,2,5,9,8]),color ='g');
plt.plot(np.arange(5,11),color ='r');
plt.legend(['1','2','3'], shadow = True, loc = 'upper left')
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize=[20,10])

#1 
plt.subplot(2,2,1)
x= [1,3,4,7,8];
plt.pie(x);
plt.title('First')

#2
plt.subplot(2,2,2)
plt.plot([1,2,3,4,5],[10,20,3,1,5]);
plt.plot(pd.DataFrame([5,2,5,9]));
plt.title('Second');

#3
plt.subplot(2,2,3)
plt.scatter(np.random.randn(20),np.random.randn(20),alpha = 0.4 , s=50);
plt.title('third');
```
