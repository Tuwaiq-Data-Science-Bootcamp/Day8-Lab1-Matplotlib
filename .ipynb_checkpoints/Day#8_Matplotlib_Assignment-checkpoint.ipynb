```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
x= np.array([1,2,3,4,5,2])
plt.plot(x);
```

```python
x= pd.DataFrame([1,2,3,4,5,2])
plt.plot(x);
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(np.arange(1,4), labels=['Jan','Septemper','Mar'], explode =[.3,.3,.3]);
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
q = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(q,y, alpha = 0.4, marker = 'o', s=50);
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
q = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(q,y, alpha = 0.4, marker = 'o', s=50)
plt.title('Scatter Chart', loc="left", color='r')
plt.xlabel('Q' , size=16)
plt.ylabel('Y', size=16);
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.plot([1,2,3,4,5], 'b')
plt.plot([10,9,8,6,3], 'g')
plt.plot([12,13,10,20,13], 'r')
plt.legend(['2020','2021','2022'], shadow= True , loc = 'left')
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize=[20,10])
plt.suptitle('Subplots',fontsize = 16)

#first
plt.subplot(1,3,1) #(number of rows, number of columns, number of plot)
plt.pie(np.arange(1,5),labels=['Jan','Feb','Mar', 'Apr']);
plt.title('Months')

#second
plt.subplot(1,3,2)
plt.plot([1,2,0,4,5])
plt.plot([10,9,8,6,3])
plt.legend(['2008','2009'], shadow = True, loc = 0)
plt.title('Years');

#Third
plt.subplot(1,3,3) #(number of rows, number of columns, number of plot)
x = np.arange(10)
y = np.arange(10)
plt.bar(x, y)
plt.xlabel('Days')
plt.ylabel('Sales')
plt.title('Sales');
```
