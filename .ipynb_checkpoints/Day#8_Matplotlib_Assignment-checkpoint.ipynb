```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python

plt.subplot(1,2,1) 
chart1=plt.plot(pd.Series([9,5,3,7]))
chart2=plt.plot(pd.DataFrame([5,10,15,20]))

plt.subplot(1,2,2) 
chart3=plt.bar(pd.Series([5,10,15,20]),pd.Series([5,10,15,20]))

print(chart1);
print(chart2);
```

## Q2: Create pie chart with equal explode values.

```python
chart3=plt.pie(np.arange(1,5), labels=['Jan','Feb','Mar', 'Apr'], explode = [.1,.1,.1,.1]);
chart3;
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x = np.random.randn(500)
y = np.random.randn(500)
chart4=plt.scatter(x, y, marker = 'o', s = 50, alpha=0.4);
chart4;
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(x, y, marker = 'o', s = 50, alpha=0.4)
plt.title('My Data', fontsize = 16, loc='left', color='r');
plt.xlabel("X",fontsize = 16) 
plt.ylabel("Y",fontsize = 16)
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.plot([2,4,8,10,12],color ='blue')
plt.plot([10,5,20,15,25],color ='m')
plt.plot([1,2,0,4,5],color ='y')
plt.legend(['A','B','C'], shadow = True, loc = 0)
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize=[20,10])
plt.suptitle('Subplots',fontsize = 16)

#first
plt.subplot(1,3,1) 
plt.pie(np.arange(1,5),labels=['Jan','Feb','Mar', 'Apr']);
plt.title('students')

#second
plt.subplot(1,3,2)
plt.plot([1,2,0,4,5])
plt.plot([10,9,8,6,3])
plt.legend(['2008','2009'], shadow = True, loc = 0)
plt.title('teachers');


#third
x=pd.Series([1,2,4,2,5,3])
y=pd.Series([10,20,40,20,50,30])

plt.subplot(1,3,3)
plt.bar(x,y)

plt.legend(['2001','2002'], shadow = True, loc = 0)
plt.title('other');
```
