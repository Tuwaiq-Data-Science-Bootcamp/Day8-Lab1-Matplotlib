```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
x = [15,20, 25, 30, 35]
plt.plot(x, [6,2,3,8,5]) , plt.plot(pd.DataFrame([1,3,4,2]));
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(np.arange(1,4), labels=['tiktok','snapchat','instagram'], explode = [.1,.1,.1]);
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x = np.random.randn(100)
y = np.random.randn(100)

plt.scatter(x, y, marker = 'o', alpha = 0.5, s = 50);
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.title('Pic', fontsize = 20, loc='left', color='r');
x = np.random.randn(200)
y = np.random.randn(200)

plt.scatter(x, y, marker = 'o', alpha = 0.5, s = 50);
plt.xlabel('X')
plt.ylabel('Y')
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.plot([1,2,3,4,5], 'g')
plt.plot([8,6,4,2,0], 'b')
plt.plot([2,5,7,9,12], 'r')
plt.legend(['Nov','Dec', 'Apr'], shadow = True, loc = 6)
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize=[20,10])
plt.suptitle('Subplots')

#2
plt.subplot(2,2,1) #(number of rows, number of columns, number of plot)
plt.plot([1,2,0,4,5])
plt.title('1')

#1
plt.subplot(2,2,2)
plt.plot([1,2,0,4,5])
plt.plot([2,7,9,6,2])
plt.title('2');

#3
plt.subplot(2,2,3)
plt.plot([1,2,0,4,5])
plt.plot([10,9,8,6,3])
plt.title('3');
```
